```python
import pandas as pd

df = pd.read_csv('airline_satisfaction.csv')
df.sample(n=10000, random_state=3)
df.head()
```

```python
df.dropna(inplace=True)
df.isna().sum() / df.shape[0]
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
pd.set_option('display.max_columns', None)

# Set X and y
y = df.satisfaction
X = df.drop(columns=['satisfaction'])
X = pd.get_dummies(X, drop_first=True) # dummy code AFTER X and y

# normalize the data
X = pd.DataFrame(MinMaxScaler().fit_transform(X), columns=X.columns) # that last part puts the column names back in

# split: train on one half and test on the other
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # putting 30% in testing
clf = DecisionTreeClassifier(random_state=3).fit(X_train, y_train) #clf stands for classification, as in classification model
print(f'Accuracy: {clf.score(X_test, y_test)}')
```
