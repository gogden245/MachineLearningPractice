```python
# Question 1: 

# Read each dataset into its own DataFrame; e.g. df_projects, df_resources, df_schools, etc.
# NOTE: donors.db contains a single table called "donors"
# print the columns found in each DataFrame in any order you want

import pandas as pd
import sqlite3
import json

# Reading csv --> DataFrame
df_projects = pd.read_csv('projects.csv')

df_resources = pd.read_csv('resources.csv')

df_schools = pd.read_csv('schools.csv')

df_teachers = pd.read_csv('teachers.csv')

# Reading json --> DataFrame
df_donations = pd.DataFrame(pd.read_json('donations.json', orient='records'))

# Reading sql --> DataFrame
conn = sqlite3.connect('donors.db')
df_donors = pd.read_sql_query("SELECT * FROM donors", conn)

#Print Columns
print(f"df_projects columns: \n {df_projects.columns} \n \n \
df_resources columns: \n {df_resources.columns} \n \n \
df_schools columns: \n {df_schools.columns} \n \n \
df_teachers columns: \n {df_teachers.columns} \n \n \
df_donations columns: \n {df_donations.columns} \n \n \
df_donors columns: \n {df_donors.columns}")
```

```python
# Question 2: 

# Merge the six DataFrames into a single DataFrame called df
# Infer the primary key/foreign key relationships based on connecting matching column ID names
# Print the complete column list of df
# Print the first five records

df = pd.merge(df_projects, df_resources, on='Project ID')
df = pd.merge(df, df_schools, on='School ID')
df = pd.merge(df, df_teachers, on='Teacher ID')
df = pd.merge(df, df_donations, on='Project ID')
df = pd.merge(df, df_donors, on='Donor ID')

pd.set_option('display.max_columns', 1000)
df.head()
```

```python
# Question 3:

# Print the number of rows in this new DataFrame df
print(df.shape[0])
```

```python
# Question 4:

# Filter out all records that are not in Utah County. In other words, only Utah County records should remain in the DataFrame df
# Print the number of rows in this reduced DataFrame

df = df[df['School County'] == 'Utah']
df.shape[0]
```

```python
# Question 5:

# Print the average donation value from this reduced dataset

print('$', df['Donation Amount'].mean())
```

```python
# Question 6:

# Print the median donation value from this reduced dataset
print('$', df['Donation Amount'].median())
```

```python
# Question 7:

# Print the donation value skewness from this reduced dataset

print('$', df['Donation Amount'].skew())
```

```python
# Question 8:

# Store df in a new sqlite db called donors_choose.db using the table name donors_choose

conn = sqlite3.connect('donors_choose.db')
cursor = conn.cursor()
df.to_sql('donors_choose', conn, index=False, if_exists='replace')
```
