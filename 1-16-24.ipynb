```python
import pandas as pd

df = pd.read_csv('insurance.csv')

df.head(10)
```

This object (below) looks like a dictionary but currently looks like a strong (you can tell because of the first single quote)

```python
import json

json_str = df.to_json(orient='columns') # this .to_json is still a pandas package
json_str
```

This (below) is how you convert it to a formal json dictionary instead of a string.

```python
df_json = json.loads(json_str)
df_json
```

Now we want to explore this dictionary

```python
print(df_json.keys())

for k, v in df_json.items(): # this helps print out all key value pairs
  print(k,v)

# create a list: []
# CREATE a dictionary: {}
# read a list/reference/index: [] -> [position index], ex [2]
# read a dictionary: [] -> ['key name']

df_json['bmi']['1']
```

```python
# a split is a list; it's faster to work with but harder to read as a human
json_str = df.to_json(orient='split') 
json_str
```

```python
df_json = json.loads(json_str)

for k, v in df_json.items(): # this helps print out all key value pairs- .items is a property of dictionaries
  print(k,v)

print('\n')
df_json['data'][1][2] # don't forget about the 0-base counting
```

```python
# this makes a LIST of DICTIONARIES
json_str = df.to_json(orient='records') 
json_str
```

```python
df_json = json.loads(json_str)
#for every item in the list, print out each item one at a time- in this case the items are dictionaries
for i in df_json:
  print(i)

print('\n')
df_json[1]['bmi']
```
