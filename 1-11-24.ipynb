```python
import pandas as pd

df=pd.read_csv('nba_salaries.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
print(df.shape)
df.head()
```

```python
def univariate(df):
  import pandas as pd
  import seaborn as sns
  import matplotlib.pyplot as plt

  # Create a daataframe to display info
  df_output = pd.DataFrame(columns=['type', 'missing', 'unique', 'min', 'q1', 'median',
                                    'q3', 'max', 'mode', 'mean', 'std', 'skew', 'kurt'])

  for col in df:
    # features that apply to all datatypes
    missing = df[col].isna().sum() # identify missing and count them up (missing is a variable)
    unique = df[col].nunique()
    mode = df[col].mode()[0]
    if pd.api.types.is_numeric_dtype(df[col]):
      #features for numeric datatypes
      min = df[col].min()
      q1 = df[col].quantile(0.25)
      median= df[col].median()
      q3 = df[col].quantile(0.75)
      max = df[col].max()
      mean = df[col].mean()
      std = df[col].std()
      skew = df[col].skew()
      kurt = df[col].kurt()
      df_output.loc[col] = ["numeric", missing, unique, min, q1, median, q3, max, mode, mean, std, skew, kurt] #if it doesn't exist, it will create it
      sns.histplot(data=df, x=col)
      plt.show()
    else:
      df_output.loc[col] = ["categorical", missing, unique, '-', '-', '-', '-', '-', mode, '-', '-', '-', '-']
      sns.countplot(data=df, x=col)
      plt.show()
  return df_output


univariate(df) #this is where we're calling the above function
```
